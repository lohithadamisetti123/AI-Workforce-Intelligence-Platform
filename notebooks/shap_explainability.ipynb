```python
import pandas as pd
import shap
import joblib
```

```python
X_train = pd.read_csv(
    '../data/X_train.csv'
)

X_test = pd.read_csv(
    '../data/X_test.csv'
)
```

```python
model = joblib.load(
    '../models/random_forest.pkl'
)
```

```python
explainer = shap.TreeExplainer(
    model
)
```

```python
shap_values = explainer.shap_values(
    X_test
)
```

```python
shap.summary_plot(
    shap_values,
    X_test
)
```

```python
employee = 10

shap.force_plot(
    explainer.expected_value[1],
    shap_values[:,:,1][employee],
    X_test.iloc[employee],
    matplotlib=True
)
```
