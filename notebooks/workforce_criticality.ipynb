```python
import pandas as pd
import joblib
```

```python
df = pd.read_csv('../data/X_test.csv')
```

```python
model = joblib.load(
    '../models/logistic_scaled.pkl'
)

scaler = joblib.load(
    '../models/scaler.pkl'
)
```

```python
df_scaled = scaler.transform(df)
```

```python
df['AttritionProbability'] = \
    model.predict_proba(df_scaled)[:,1]
```

```python
df['AttritionProbability'].head()
```

```python
df['CriticalityScore'] = (
    df['AttritionProbability']
    *
    df['EmployeeValueScore']
    *
    df['ReplacementDifficulty']
    *
    (df['FinancialLoss']/100000)
)
```

```python
def classify(score):

    if score >= 31588:
        return "Critical"

    elif score >= 12271:
        return "Medium"

    else:
        return "Low"


df['CriticalityLevel'] = \
    df['CriticalityScore'].apply(classify)
```

```python
def priority(row):

    if row['CriticalityLevel'] == "Critical":
        return "Immediate Action"

    elif row['CriticalityLevel'] == "Medium":
        return "Review Within 30 Days"

    return "Monitor"


df['ActionPriority'] = df.apply(
    priority,
    axis=1
)
```
