```python
import pandas as pd

from imblearn.over_sampling import SMOTE
```

```python
X_train = pd.read_csv('../data/X_train.csv')

y_train = pd.read_csv('../data/y_train.csv')
```

```python
y_train = y_train.values.ravel()
```

```python
smote = SMOTE(
    random_state=42
)

X_smote, y_smote = smote.fit_resample(
    X_train,
    y_train
)
```

```python
print(y_train.shape)

print(y_smote.shape)
```

```python
pd.Series(y_smote).value_counts()
```

```python
X_smote.to_csv(
    '../data/X_smote.csv',
    index=False
)

pd.DataFrame(y_smote).to_csv(
    '../data/y_smote.csv',
    index=False
)
```
