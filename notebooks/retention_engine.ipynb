```python
import pandas as pd
import joblib
```

```python
df = pd.read_csv(
    '../data/workforce_criticality.csv'
)

df.head()
```

```python
def recommend(row):

    rec = []

    if row['JobSatisfaction'] <= 2:
        rec.append(
            "Improve job satisfaction"
        )

    if row['WorkLifeBalance'] <= 2:
        rec.append(
            "Provide flexible work policy"
        )

    if row['StockOptionLevel'] == 0:
        rec.append(
            "Offer stock incentives"
        )

    if row['JobInvolvement'] <= 2:
        rec.append(
            "Increase employee engagement"
        )

    if row['EnvironmentSatisfaction'] <= 2:
        rec.append(
            "Improve work environment"
        )

    if row['OverTime'] == 1:
        rec.append(
            "Reduce overtime workload"
        )

    if row['MonthlyIncome'] < 5000:
        rec.append(
            "Consider salary adjustment"
        )

    if row['YearsSinceLastPromotion'] > 5:
        rec.append(
            "Provide promotion opportunity"
        )

    if len(rec) == 0:
        rec.append(
            "Maintain current retention strategy"
        )

    return rec
```

```python
df['Recommendations'] = \
    df.apply(
        recommend,
        axis=1
    )
```

```python
df[
    [
        'AttritionProbability',
        'CriticalityLevel',
        'Recommendations'
    ]
].head(20)
```

```python
df.to_csv(
    '../data/retention_recommendations.csv',
    index=False
)
```
