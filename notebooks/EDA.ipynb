```python
categorical=df.select_dtypes(include='object')

categorical.columns
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Department',
    hue='Attrition'
)

plt.xticks(rotation=20)
plt.show()
```

Identify departments with the highest employee turnover.

```python
plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x='OverTime',
    hue='Attrition'
)

plt.show()
```

Employees performing overtime show significantly higher attrition.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Attrition',
    y='MonthlyIncome'
)

plt.show()
```

Lower income employees demonstrate higher attrition tendencies.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='JobSatisfaction',
    hue='Attrition'
)

plt.show()
```

Lower job satisfaction correlates strongly with attrition.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='WorkLifeBalance',
    hue='Attrition'
)

plt.show()
```

Poor work-life balance increases employee exit probability.

```python
plt.figure(figsize=(10,5))

sns.histplot(
    data=df,
    x='YearsAtCompany',
    hue='Attrition',
    kde=True
)

plt.show()
```
